## `if .. else .. else if ..`

````python
%%writefile conditionals.cpp
# include <iostream>

int x = 10;
int y = 20;

int main()
{
    if (x < y)
    {
        std::cout << "x is smaller than y" << std::endl;
    }
    else if (x == y)
    {
        std::cout << "x equal to y" << std::endl;
    }
    else
    {
        std::cout << "x is bigger than y" << std::endl;
    }
}
````

````python
%%shell

g++ conditionals.cpp -o output1
./output1
````

### The `?` Operator
`?` operator can be use to replace `if .. else ..` statement.
```
Val1 ? Val2 : Val3
```
The value of a `?` expression is determined like this: `Val1` is evaluated. If it is `true`, then `Val2` is evaluated and becomes the value of the entire `?` expression. If `Val1` is `false`, then `Val3` is evaluated and its value becomes the value of the expression.

## `switch`
A `switch` statement allows a variable to be tested for equality against a list of values. Each value is called a case, and the variable being switched on is checked for each case.

![image.png](https://www.tutorialspoint.com/cplusplus/images/cpp_switch_statement.jpg)

The following rules apply to a switch statement âˆ’

1. The `expression` used in a `switch` statement must have an integral or enumerated type, or be of a class type in which the class has a single conversion function to an integral or enumerated type.
2. You can have any number of case statements within a `switch`. Each case is followed by the value to be compared to and a colon.
3. The `constant-expression` for a `case` must be the same data type as the variable in the switch, and it must be a constant or a literal. For `string` or `char` value, constant was wrapped with `'`.
4. When the variable being switched on is equal to a case, the statements following that case will execute until a `break` statement is reached.
5. When a `break` statement is reached, the `switch` terminates, and the flow of control jumps to the **next line** following the `switch` statement.
6. Not every `case` needs to contain a `break`. If no `break` appears, the flow of control will fall through to subsequent cases until a `break` is reached.
7. A `switch` statement can have an optional `default` case, which must appear at the end of the `switch`. The default case can be used for performing a task when none of the cases is true. No `break` is needed in the `default` case.

````python
%%writefile switch.cpp
# include <iostream>

int main()
{
    char grade = 'D';
    
    switch(grade)
    {
        case 'A':
            std::cout << "Great!";
            break;
        case 'B':
            std::cout << "Well done!";
            break;
        case 'C':
            std::cout << "Hmm... not bad";
            break;
        case 'D':
            std::cout << "You better do better next!";
            break;
        case 'F':
            std::cout << "You failed, sorry";
            break;
        default:
            std::cout << "Invalid grade";
    }
 
    std::cout << " Your grade is " << grade << std::endl;
    return 0;
}
````
