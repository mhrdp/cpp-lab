# Create CPP Script
One way to run CPP / C++ script inside Google Colab is by utilizing `magic command` that prefixed by `%` or `%%`.

In this case we will use `%%writefile <filename.cpp>` to write a binary C++ file, and then we will read the file using `%%shell` and `G++` compiler (a compiler for C++)

You can check the available `magic command` using `%Ismagic` and run the block.

## This script below is a simple example for C++ script on Colab
When you run the block, it will write the `.cpp` file but will not return any result whatsoever.

```python
%%writefile SimpleCPPOperations.cpp
#include <iostream>

int
    main()
{
    std::cout << "Welcome To GeeksforGeeks\n";
    return 0;
}
```

## To run the the written `.cpp` script above:

Essentially, this will tell the `shell` to run `SimpleCPPOperations.cpp` with `G++` compiler, and save its output as `output` before accessing it by `./output`.

```python
%%shell

g++ SimpleCPPOperations.cpp -o output
./output
```
